- ### Find the Most Popular Trios - I

If a producer is looking to make a blockbuster movie. There will primarily be three lead roles in your movie and you wish to cast the most popular actors for it. Now, since we don't want to take a risk, we will cast a trio which has already acted in together in a movie before. The metric that we've chosen to check the popularity is the Facebook likes of each of these actors.

```python
# Write your code here

trio = movies[['actor_1_name','actor_2_name','actor_3_name','actor_1_facebook_likes','actor_2_facebook_likes','actor_3_facebook_likes']]
trio
```

```python
# Creating a new column - total_likes which will have sum of all likes of the trio of actors
trio['total_likes']=trio[['actor_1_facebook_likes','actor_2_facebook_likes','actor_3_facebook_likes']].sum(axis=1)
trio
```

```python
#This is the answer to the Subtask 2.4: (Output of Most popular Trios-1 in a list)

top5_trio = trio.sort_values(by='total_likes', ascending=False).head(5)
top5_trio[['actor_1_name','actor_2_name','actor_3_name']].values.tolist()
```

- ### Find the Most Popular Trios - II

In the previous segment of code, we found the popular trio based on the total number of facebook likes. Let's add a small condition to it and make sure that all three actors are popular. The condition is **none of the three actors' Facebook likes should be less than half of the other two**. 

Having this condition ensures that we aren't getting any unpopular actor in your trio (since the total likes calculated in the previous question doesn't tell anything about the individual popularities of each actor in the trio.).

```python
# Your answer here (optional and not graded)

def likes(x):
    l1 = x['actor_1_facebook_likes']
    l2 = x['actor_2_facebook_likes']
    l3 = x['actor_3_facebook_likes']
    
    Flag= True
    min_like = min(l1,l2,l3)
    
    if min_like == l1:
        if min_like <l2/2 and min_like<l3/2:
            Flag = False
    elif min_like == l2:
        if min_like <l3/1 and min_like<l1/2:
            Flag = False
    elif min_like == l3:
        if min_like <l2/2 and min_like<l1/2:
            Flag = False
    return Flag
```

```python
top5_trio['most_popular'] = top5_trio.apply(likes, axis=1)
top5_trio
```

- **`No. of trios that satisfy the above condition:`** There are 3 Trios which satisfy the above condition

- **`Most popular trio after applying the condition:`** [[Leonardo DiCaprio,Tom Hardy,Joseph Gordon-Levitt],[Jennifer Lawrence,Peter Dinklage,Hugh Jackman],[Tom Hardy,Christian Bale,Joseph Gordon-Levitt]]

## Demographic analysis

If we take a look at the last columns in the dataframe, most of these are related to demographics of the voters. We also have three genre columns indicating the genres of a particular movie. We will extensively use these columns for the third and the final stage of our analysis wherein we will analyse the voters across all demographics and also see how these vary across various genres.

-  ###  Combine the Dataframe by Genres


```python
# Create the dataframe df_by_genre

df_by_genre = movies[['genre_1','genre_2','genre_3','CVotes10',\
                     'CVotes09','CVotes08','CVotes07','CVotes06','CVotes05','CVotes04',\
                     'CVotes03','CVotes02', 'CVotes01', 'CVotesMale', 'CVotesFemale',\
                     'CVotesU18', 'CVotesU18M', 'CVotesU18F', 'CVotes1829', 'CVotes1829M',\
                     'CVotes1829F', 'CVotes3044', 'CVotes3044M', 'CVotes3044F', 'CVotes45A',\
                     'CVotes45AM', 'CVotes45AF', 'CVotes1000', 'CVotesUS', 'CVotesnUS',\
                     'VotesM', 'VotesF', 'VotesU18', 'VotesU18M', 'VotesU18F', 'Votes1829',\
                     'Votes1829M', 'Votes1829F', 'Votes3044', 'Votes3044M', 'Votes3044F',\
                     'Votes45A', 'Votes45AM', 'Votes45AF', 'Votes1000', 'VotesUS','VotesnUS']]
df_by_genre.shape
```

```python
# Create a column cnt and initialize it to 1

df_by_genre['cnt'] = 1
```

```python
# Group the movies by individual genres

df_by_g1 = df_by_genre.groupby(by='genre_1').sum()
df_by_g2 = df_by_genre.groupby(by='genre_2').sum()
df_by_g3 = df_by_genre.groupby(by='genre_3').sum()
```

```python
# Add the grouped data frames and store it in a new data frame

df_add = df_by_g1.add(df_by_g2, fill_value=0).add(df_by_g3, fill_value=0)
df_add
```

```python
# Extract genres with atleast 10 occurences

genre_top10 = df_add[df_add['cnt']>=10]
genre_top10
```

```python
# Take the mean for every column by dividing with cnt 
temp1 = genre_top10['cnt']
genre_top10 = genre_top10.loc[:,:'VotesnUS'].div(genre_top10['cnt'], axis=0)
genre_top10['cnt'] = temp1
genre_top10
```

```python
# Rounding off the columns of Votes to two decimals

genre_top10 = genre_top10.round(2)
```

```python
# Converting CVotes to int type
cols = ['CVotes10','CVotes09','CVotes08','CVotes07','CVotes06','CVotes05','CVotes04',\
                     'CVotes03','CVotes02', 'CVotes01', 'CVotesMale', 'CVotesFemale',\
                     'CVotesU18', 'CVotesU18M', 'CVotesU18F', 'CVotes1829', 'CVotes1829M',\
                     'CVotes1829F', 'CVotes3044', 'CVotes3044M', 'CVotes3044F', 'CVotes45A',\
                     'CVotes45AM', 'CVotes45AF', 'CVotes1000', 'CVotesUS', 'CVotesnUS']
genre_top10[cols] = genre_top10[cols].astype(int)
genre_top10
```

If we take a look at the final dataframe that we have here, we will see that we now have the complete information about all the demographic (Votes- and CVotes-related) columns across the top 10 genres. We can use this dataset to extract exciting insights about the voters!

-  ###  US vs non-US Cross Analysis

The dataset contains both the US and non-US movies. Let's analyse how both the US and the non-US voters have responded to the US and the non-US movies.

**`Inferences:`** Write your two inferences/observations below:
- Inference 1: It seems that US movies have more voters from both US and non-US origin as compared to non-US movies.
- Inference 2: It can also be inferred from the boxplot of 'IFUS vs CVotesnUS' that, the 50th percentile or 2nd Quantile value is almost same for both US and non-US movies for all non-US votes.

```python
# Box plot - 2: VotesUS(y) vs IFUS(x)

fig, axis = plt.subplots(1,2, figsize=[10,6])
sns.boxplot(data=movies, y='VotesUS',x='IFUS', ax=axis[0])
axis[0].set_title('VotesUS vs IFUS', fontdict={'fontsize':17,'color':'Purple'})
sns.boxplot(data=movies, y='VotesnUS', x='IFUS', ax=axis[1])
axis[1].set_title('VotesnUS vs IFUS', fontdict={'fontsize':17,'color':'Purple'})
plt.show()
```
